Import The Libraries

```python
from ast import increment_lineno
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

Loading the Dataset

```python
#load the data
iris= pd.read_csv('IRIS.csv')
iris.head(50)
```

Virtualization of Dataset


```python
iris.describe()
```

```python
iris["Species"].value_counts()
```

Scatter Plot

```python
sns.FacetGrid(iris, hue="Species",height=6).map(plt.scatter, "Petal.Length" , "Sepal.Width").add_legend()
```

Logistic Regression

```python
flower_mapping = {'setosa': 0, 'versicolor':1, 'virginica':2}
iris["Species"] = iris["Species"].map(flower_mapping)
```

```python
iris.head()
```

Preparing inputs and outputs

```python
x=iris[['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']].values
y=iris[['Species']].values
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x,y)
```

Accuracy

```python
model.score(x,y)
```

Make Predictions

```python
expected = y
predicted = model.predict(x)
predicted
```

Summarize the fit of the model

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(expected,predicted))
```

```python
print(metrics.confusion_matrix(expected,predicted))
```

Regularization

```python
model = LogisticRegression(C=20,penalty='l2')
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```
